```python
import pandas as pd
import numpy as np
```

```python
station_number = 3904
file_prefix = 'hly'
station_url = f'http://cli.fusio.net/cli/climate_data/webdata/{file_prefix}{station_number}.csv'
```

```python
df_tmp = pd.read_csv(station_url,nrows=50,sep='$')
n_skiprows = np.min(df_tmp[df_tmp[df_tmp.columns[0]].str[:8]=='date,ind'].index.values)
n_blanks = 3
```

```python
df_header = pd.read_csv(station_url,nrows=n_skiprows,sep='$')
df_header.to_csv(f'{file_prefix}{station_number}_header.csv',index=False)
```

```python
df = pd.read_csv(station_url,skiprows=n_skiprows+n_blanks)
df = df.set_index('date')
df.head()
```

```python
df = df.apply(pd.to_numeric,errors='coerce')
```

```python
df.dtypes
```

```python
df.to_parquet(f'{station_number}.parquet')
```

```python
df.to_csv(f'{station_number}.csv')
```
