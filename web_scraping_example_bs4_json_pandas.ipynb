```python
import pandas as pd
import bs4 as bs
import urllib.request
```

```python
# mess around with header to get the data
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
url = "https://auburn.rivals.com/commitments/football/2018"
headers={'User-Agent':user_agent,} 
request=urllib.request.Request(url,None,headers) #The assembled request

# get response
response = urllib.request.urlopen(request)

# get raw source data of the page
data = response.read()

# parse it into a bs4 object
soup=bs.BeautifulSoup(data, "html5lib")
```

```python
# get specific tag that holds the data inside it in json
mydata = soup.find('rv-commitments')['prospects']

# read that json into a df
df = pd.read_json(mydata)

# here is your df
df
```
