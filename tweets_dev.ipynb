```python
import json
import pandas as pd
import os

print(os.getcwd())
```

```python
tweets = []
for line in open('tweets.json', 'r'):
    tweets.append(json.loads(line))
```

```python
df_tweets = pd.DataFrame.from_dict(tweets)
print(df_tweets.shape)
df_tweets.head(2)
```

```python
# just normalise the nested json col on its own into a df
df_retweeted_status = pd.DataFrame(df_tweets['retweeted_status'].to_dict()).transpose()
# add in id based on common index position
df_retweeted_status = df_tweets[['id']].merge(df_retweeted_status,'inner',left_index=True,right_index=True,suffixes=('_df_tweets', ''))

# look at data
print(df_retweeted_status.shape)
df_retweeted_status.head()
```
