```python
import pandas as pd
import json
from textblob import TextBlob
from textblob import Blobber
import timeit
```

```python
business_small = pd.DataFrame.from_csv('yelp_academic_dataset_business_small.csv', index_col=None)
review_small = pd.DataFrame.from_csv('yelp_academic_dataset_review_small.csv', index_col=None, encoding = "ISO-8859-1")
review_tiny = review_small.head(20)
```

```python
# function that loops though each row
def process_reviews():
    polarity = []
    subjectivity = [] 
    # loop through each review
    for x in range(len(review_tiny)): 
        # get polarity and subjectivity
        pol, subj = TextBlob(review_tiny.text[x]).sentiment
        # append to list
        polarity.append(pol)
        subjectivity.append(subj)
        
    return(polarity,subjectivity)

#mypolarity,mysubjectivity = process_reviews(review_small)

ntrials = 50

print(timeit.timeit(process_reviews, number=ntrials)/ntrials)
```

```python
# function that uses map()
def process_reviews_fast():

    review_list = list(review_tiny['text'].astype(str))
    
    review_list_blobs = map(TextBlob, review_list)
    df_sentiment = pd.DataFrame([blob.sentiment for blob in review_list_blobs])
            
    return(df_sentiment)

ntrials = 50

print(timeit.timeit(process_reviews_fast, number=ntrials)/ntrials)
```

```python
# using existing function
mypolarity, mysubjectivity = process_reviews()
# add results into df
rev_sentiment =  pd.DataFrame({'review_id': review_tiny.review_id, 'polarity': mypolarity, 'subjectivity': mysubjectivity })
review_tiny_final = review_tiny.merge(rev_sentiment)
review_tiny_final[['polarity', 'subjectivity']]
```

```python
# using new function
df_sentiment = process_reviews_fast()
df_sentiment['review_id'] = review_tiny['review_id']
# add results into df
rev_sentiment_fast =  df_sentiment
review_tiny_final_fast = review_tiny.merge(rev_sentiment_fast)
review_tiny_final_fast[['polarity', 'subjectivity']]
```
